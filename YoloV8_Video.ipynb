```python
from ultralytics import YOLO
```

```python
import matplotlib.pyplot as plt
import cv2
```

```python
model = YOLO('C:/Users/HP/YoloV8_Large/runs/detect/train7/weights/best.pt')
```

```python
cap = cv2.VideoCapture("Video_test8.mp4")
# cap.set(3,640) # adjust width
# cap.set(4,480) # adjust height
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
```

```python
vid_cod = cv2.VideoWriter_fourcc(*'mp4v')
output = cv2.VideoWriter("cam_video.mp4", vid_cod, 30.0, (frame_width,frame_height))
```

```python
while cap.isOpened():
    success, img = cap.read()
    if not success:
        break
    # Doing detections using YOLOv8 frame by frame
    #stream = True will use the generator and it is more efficient than normal
    results=model(img,save=True)
    print(results)
    cv2.waitKey(1)
    res_plotted = results[0].plot()
    cv2.imshow("result",res_plotted)
    output.write(res_plotted)
    if cv2.waitKey(1) & 0xFF==ord('q'): # quit when 'q' is pressed
        cap.release()
        break

cap.release()
output.release()
cv2.destroyAllWindows() 
```
